```python
df.drop_duplicates(inplace= True)
```

```python
df.__len__()
```

```python
lst_column = []
for i in df.columns :
    lst_column.append(i.replace("[", "").replace("]", "").strip())
lst_column[0] = "Period\\Unit"
lst_column
```

```python
df.columns = lst_column
```

```python
df.head()
```

```python
df.to_csv("CLEANED_euro_daily_hist_1999-2023.csv",index= False)
```
