```python
lst = []
for i in df["Holiday"] :
    if isna(i) :
        continue
    lst.append(i)
lst
```

```python
# finding the column with nan values ...
lst_column_with_nan_values = []
for i in df.columns :
    if df[isna(df[i])].__len__() != 0 :
        lst_column_with_nan_values.append(i)
lst_column_with_nan_values
```

```python
lst = []
for i in df["Holiday"] :
    if(isna(i)) :
        lst.append("-")
    else :
        lst.append(i.capitalize())
lst
```

```python
df["Holiday"] = lst
```

```python
df.head()
```

```python
# converting the Temp column values to Temp in Kelvin, Fahrenheit and Celsius

def kelvin_to_celsius(kelvin : float) -> float :
    return round(kelvin - 273.15,3)

def kelvin_to_fahrenheit(kelvin: float) -> float:
    return round(kelvin * 1.8 - 459.67 , 3)
```

```python
temp_celsius = []
temp_fahrenheit = []

for i in df["Temp"] :
    temp_celsius.append(kelvin_to_celsius(float(i)) )
    temp_fahrenheit.append(kelvin_to_fahrenheit(float(i)) )
```

```python
temp_celsius
```

```python
df["Temp(in C)"] = temp_celsius
df["Temp(in F)"] = temp_fahrenheit
```

```python
df.head()
```

```python
lst_column = list(df.columns )[:-2]
lst_column.insert(2,"Temp(in C)")
lst_column.insert(3,"Temp(in F)")
lst_column
```

```python
df = df.reindex(columns= lst_column )
```

```python
lst_date = []
lst_time = []
lst_exception_rows = []
for i in df["Date_time"] :
    try :
        x = str(i).strip().split(" ")
        lst_date.append(x[0])
        lst_time.append(x[1])
    except :
        lst_exception_rows.append(i)
lst_exception_rows
```
