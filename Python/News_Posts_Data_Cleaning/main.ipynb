```python
from pandas import read_csv, isna
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.__len__()
```

```python
lst_columns = []
for i in df.columns :
    lst_columns.append(i.capitalize())
lst_columns
```

```python
df.columns = lst_columns
```

```python
df.head()
```

```python
# finding the columns with Nan Values in it :
lst_column_with_nan_values  = []
for i in df.columns :
    if df[isna(df[i])].__len__() != 0 :
        lst_column_with_nan_values.append(i)
lst_column_with_nan_values
```

```python
lst_exception_rows = []
lst = []
for i in df["Author"] :
    lst.append(i.capitalize())
lst_exception_rows
```

```python
df_date = []
df_month = []
df_year = []
df_time = []
lst_exception_rows = []
for i in df["Created_at"] :
    try :
        temp = i.split(" ")
        df_time.append(temp[1])
        temp = temp[0].split("/")
        df_month.append(temp[0])
        df_date.append(temp[1] )
        df_year.append(temp[2])
    except :
        lst_exception_rows.append(i)
lst_exception_rows
```

```python
len(df_date) == len(df_month) == len(df_year) == len(df_month) == len(df)
```

```python
df["Day"] = df_date
df["Month"] = df_month
df["Year"] = df_year
df["Time"] = df_time
```
